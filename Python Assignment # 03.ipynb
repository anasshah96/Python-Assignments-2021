# Question # 07

```python
sub1 = int(input("Math = "))

sub2 = int(input("English = "))

sub3 = int(input("Physics = "))

sub4 = int(input("Chemistry = "))

sub5 = int(input("Computer Studies = "))

totalMarks = 500
 
if sub1 >= 0 and sub1 <= 100 and sub2 >= 0 and sub2 <= 100 and sub3 >= 0 and sub3 <= 100 and sub4 >= 0 and sub4 <= 100 and sub5 >= 0 and sub5 <= 100 :
    totalObt = (sub1 + sub2 + sub3 + sub4 + sub5)
    per = 100/totalMarks*totalObt
    print(str( totalObt) + " Marks out of " + str(totalMarks))
    print("Percentage: " + str(per) + "%")
    
    if per >= 80 and per <= 100 :
        print("Grade: 'A+'")
    elif per >= 70 and per < 80 :
        print("Grade: 'A'")
    elif per >= 60 and per < 70 :
        print("Grade: 'B'")
    elif per >= 50 and per < 60 :
        print("Grade: 'C'")
    elif per >= 40 and per < 50 :
        print("Grade: 'D'")
    elif per >= 33 and per < 40 :
        print("Grade: 'E'")
    elif per >= 0 and per < 33 :
        print("Grade: 'Fail'")
        
else:
    print("Each subject marks should be between 0 - 100!")
```
